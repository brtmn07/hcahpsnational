```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hcahps2017 = pd.read_csv('../hcahpsnational/data/hcahps_2018.csv')
```

```python
hcahps2017
```

```python
def assign_region(state):
    if state in ['CT','ME','MA','NH','RI','VT','NJ','NY','PA','MD','DE']:
        return 'Northeast'
    elif state in ['IL','IN','MI','OH','WI','IA','KS','MN','MO','NE','ND','SD','OK']:
        return 'Midwest'
    elif state in ['FL','GA','NC','SC','VA','AL','KY','MS','TN','AR','LA','TX','WV']:
        return 'South'
    elif state in ['AZ','CO','ID','MT','NV','NM','UT','WY','AK','CA','HI','OR','WA']:
        return 'West'
    else:
        return 'Other'

hcahps2017['region'] = hcahps2017['State'].apply(assign_region)
```

```python
hcahps2017
```

## making all the columns lowercase for easier usability

```python
hcahps2017.columns = hcahps2017.columns.str.lower()
```

```python
hcahps2017.head(2)
```

```python
hcahps2017.loc[:, 'year'] = pd.to_datetime(hcahps2017['measure end date'], format="%m/%d/%Y").dt.year
```

```python
hcahps2017.columns
```

```python
hcahps2017 = hcahps2017.drop(['county name', 'hospital name', 'city', 'address','phone number', 'measure start date', 'measure end date' ], axis=1)
```

## filtering all rows except star ratinngs

```python
allstarrating2017 = []
for index, row in hcahps2017.iterrows():
    if 'STAR_RATING' in row['hcahps measure id']:
        allstarrating2017.append(row)
allstarrating2017 = pd.DataFrame(allstarrating2017)
```

```python
allstarrating2017.reset_index(drop=True)
```

```python
allstarrating2017 = allstarrating2017.loc[allstarrating2017['patient survey star rating'] != 'Not Available']
```

```python
allstarrating2017
```

```python
allstarrating2017.to_csv('allstarrating2017.csv', index = False)
```

```python
allstarrating2017['hcahps question'].unique()
```
